**TCP throughput over the wired and wireless channels**

```python
import matplotlib.pyplot as plt
import numpy as np
   
txqueuelen = ['1', '10', '100', '1000', '10000']
wired_tcp = [148, 154, 159, 150, 153]
wireless_tcp = [69.9, 69.1, 66.7, 69.1, 71.1]

x = np.arange(len(txqueuelen))
width = 0.2

fig, ax = plt.subplots()

rects1 = ax.bar(x-width, wired_tcp, width, label = 'TCP wired throughput', color = 'r')
rects2 = ax.bar(x, wireless_tcp , width, label = 'TCP wireless throughput', color = 'g')

ax.set_ylabel('Throughput(Mbits/s)', fontsize = 16)
ax.set_xlabel('txqueuelen', fontsize = 16)
ax.set_title('TCP throughput of wired and wireless connections with different txqueuelen')
ax.set_xticks(x)
ax.set_xticklabels(txqueuelen)
ax.legend(loc=2, bbox_to_anchor=(1.05, 1.0), borderaxespad = 0.)
```

**UDP throughput over the wired and wireless channels**

```python
import matplotlib.pyplot as plt
import numpy as np
   
txqueuelen = ['1', '10', '100', '1000', '10000', '100000']
wired_udp_sender = [2740, 2900, 2910, 2900, 2980, 2920]
wired_udp_receiver = [435, 434, 408, 488, 472, 496]
wireless_udp_sender = [2690, 2660, 2630, 2720, 2680, 2670]
wireless_udp_receiver = [85.8, 81.0, 86.6, 84.7, 88.3, 87.4]
x = np.arange(len(txqueuelen))
width = 0.2

fig, ax = plt.subplots()

rects1 = ax.bar(x - 2*width, wired_udp_sender, width, label = 'UDP wired throughput sender', color = 'r')
rects2 = ax.bar(x - width, wired_udp_receiver , width, label = 'UDP wired throughput receiver', color = 'r', hatch='++')
rects3 = ax.bar(x, wireless_udp_sender, width, label = 'UDP wireless throughput sender', color = 'g')
rects4 = ax.bar(x + width, wireless_udp_receiver, width, label = 'UDP wireless throughput receiver', color = 'g', hatch='++')

ax.set_ylabel('Throughput(Mbits/s)', fontsize = 16)
ax.set_xlabel('txqueuelen', fontsize = 16)
ax.set_title('UDP throughput of wired and wireless connections with different txqueuelen')
ax.set_xticks(x)
ax.set_xticklabels(txqueuelen)
ax.legend(loc=2, bbox_to_anchor=(1.05, 1.0), borderaxespad = 0.)
```

**Throughput changes adjusting BQL**

```python
import matplotlib.pyplot as plt
import numpy as np

limit_max = ['0', '10', '100', '1000', '10000']
sender = [2810, 2800, 2800, 2690, 2910]
receiver = [367, 439, 467, 496, 519]

x = np.arange(len(limit_max))
width = 0.2

fig, ax = plt.subplots()

rects1 = ax.bar(x - width, sender, width, label = 'UDP wired sender', color = 'r')
rects2 = ax.bar(x , receiver , width, label = 'UDP wired receiver', color = 'g')


ax.set_ylabel('Throughput(Mbits/s)', fontsize = 16)
ax.set_xlabel('limit_max', fontsize = 16)
ax.set_title('UDP throughput with different BQL limits')
ax.set_xticks(x)
ax.set_xticklabels(limit_max)
ax.legend(loc=2, bbox_to_anchor=(1.05, 1.0), borderaxespad = 0.)

for a,b in zip(x, receiver):
    plt.text(a, b, b, ha='center', va='bottom', fontsize=10)
plt.plot(x, receiver, "g", marker='.', ms=5, linewidth = '1')
```

**Packet delay while manipulating the queue size in Linux**

From local to AWS EC2 instance 

```python
import matplotlib.pyplot as plt
import numpy as np

avg = [77.948, 77.941, 77.992, 78.015, 78.239, 78.388, 78.523]
dev = [0.655, 0.413, 0.501, 0.697, 0.966, 1.353, 1.252]
txqueuelen = ['0', '10', '100', '1000', '10000', '100000', '1000000']
x = np.arange(len(txqueuelen))
error_params=dict(elinewidth=4,ecolor='coral',capsize=5)
fig, ax = plt.subplots()

ax.set_ylabel('Delay(ms)', fontsize = 16)
ax.set_xlabel('txqueuelen', fontsize = 16)
ax.set_title('UDP packet delay with different transmission queue length (to AWS)')
ax.set_xticks(x)


plt.bar(x,avg,color='b',yerr=dev,error_kw=error_params,tick_label=['0', '10', '100', '1000', '10000', '100000', '1000000'])
plt.ylim((76, 80))

for a,b in zip(x, avg):
    plt.text(a, b, b, ha='center', va='bottom', fontsize=10)
plt.plot(x, avg, "g", marker='.', ms=5, linewidth = '1')
```
